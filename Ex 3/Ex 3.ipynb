```python
# Import libraries
import pandas as pd
df = pd.read_csv("medical_dataset_1100_rows_v2.csv")
print(df)
print(df.head())
print(df.info())
print(df.describe())
```

```python
import pandas as pd


print("===== INFO =====")
df.info()

print("\n===== TOTAL NULL VALUES =====")
print(df.isnull().sum())

print("\n===== TOTAL NA VALUES =====")
print(df.isna().sum())


print("\n===== NULL VALUES =====")
print(df.isnull())


print("\n===== NA VALUES =====")
print(df.isna())


print("\n===== NOT NULL VALUES =====")
print(df.notnull())


print("\n===== NON-NULL COUNT =====")
print(df.notnull().sum())

print("\n===== TOTAL MISSING VALUES =====")
print(df.isnull().sum().sum())


print("\n===== ANY MISSING VALUES? =====")
print(df.isnull().values.any())
```

```python
import pandas as pd


df_constant = df.fillna(0)
print("===== CONSTANT VALUE =====")
print(df_constant)


df_mean = df.copy()
df_mean["Age"] = df_mean["Age"].fillna(df_mean["Age"].mean())

print("\n===== MEAN =====")
print(df_mean)


df_median = df.copy()
df_median["Age"] = df_median["Age"].fillna(df_median["Age"].median())

print("\n===== MEDIAN =====")
print(df_median)


df_mode = df.copy()
df_mode["Disease"] = df_mode["Disease"].fillna(df_mode["Disease"].mode()[0])

print("\n===== MODE =====")
print(df_mode)


df_ffill = df.ffill()

print("\n===== FORWARD FILL =====")
print(df_ffill)


df_bfill = df.bfill()

print("\n===== BACKWARD FILL =====")
print(df_bfill)
```

```python
import pandas as pd


df_drop_rows = df.dropna()
print("===== DROP ROWS =====")
print(df_drop_rows)


df_drop_cols = df.dropna(axis=1)
print("\n===== DROP COLUMNS =====")
print(df_drop_cols)


df_any = df.dropna(how='any')
print("\n===== DROP HOW='ANY' =====")
print(df_any)


df_all = df.dropna(how='all')
print("\n===== DROP HOW='ALL' =====")
print(df_all)


df_thresh = df.dropna(thresh=5)
print("\n===== THRESH=5 =====")
print(df_thresh)


df_subset = df.dropna(subset=['Age', 'Disease'])
print("\n===== SUBSET (Age, Disease) =====")
print(df_subset)

df.dropna(inplace=True)

print("\n===== AFTER inplace=True =====")
print(df)
```
